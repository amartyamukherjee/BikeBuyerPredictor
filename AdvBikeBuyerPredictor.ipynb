```python
num_cols = ['YearlyIncome']
df_scaled = scale(df[num_cols])
df_scaled = pd.DataFrame(df_scaled, columns = num_cols)
print(df_scaled.describe().round(3))
```

```python
countries = {'Australia':0,'United States':1,'Canada':2,'Germany':3,'United Kingdom':4,'France':5}
df_scaled['CountryRegionName'] = df['CountryRegionName']
df_scaled['CountryRegionName'] = [countries[x] for x in df_scaled['CountryRegionName']]
```

```python
occupations = {'Professional':0,'Management':1,'Skilled Manual':2,'Clerical':3,'Manual':4}
df_scaled['Occupation'] = df['Occupation']
df_scaled['Occupation'] = [occupations[x] for x in df_scaled['Occupation']]
```

```python
educations = {'Bachelors ':0,'Partial College':1,'High School Education':2, 'High School':2,'Partial High School':3,'Graduate Degree':4}
df_scaled['Education'] = df['Education']
df_scaled['Education'] = [educations[x] for x in df_scaled['Education']]
```

```python
genders = {'M':0,'F':1}
df_scaled['Gender'] = df['Gender']
df_scaled['Gender'] = [genders[x] for x in df_scaled['Gender']]
```

```python
marital_status = {'M':0,'S':1}
df_scaled['MaritalStatus'] = df['MaritalStatus']
df_scaled['MaritalStatus'] = [marital_status[x] for x in df_scaled['MaritalStatus']]
```

```python
df_scaled['TotalChildren'] = df['TotalChildren']
df_scaled['NumberCarsOwned'] = df['NumberCarsOwned']
df_scaled['HomeOwnerFlag'] = df['HomeOwnerFlag']
df_scaled['NumberChildrenAtHome'] = df['NumberChildrenAtHome']
df_scaled['AveMonthSpend'] = df['AveMonthSpend']
```

```python
df_scaled['BikeBuyer']=df['BikeBuyer']
```

```python
df_scaled
```

```python
test_sample_count = 1000
count = len(df_scaled['YearlyIncome']) - test_sample_count
column_count = 12
df_train = np.asmatrix(df_scaled)[:count]
df_train_features = df_train[:,:column_count-1]
df_train_labels = np.ravel(df_train[:,column_count-1])
```

```python
df_train
```

```python
df_train_features
```

```python
df_train_labels
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN_mod = KNeighborsClassifier(n_neighbors = 2)
KNN_mod.fit(df_train_features, df_train_labels)
```

```python
df_test = df.iloc[:][count:]
df_test = df_test.reset_index()
```

```python
df_test
```

```python
num_cols = ['YearlyIncome']
df_test_scaled = scale(df_test[num_cols])
df_test_scaled = pd.DataFrame(df_test_scaled, columns = num_cols)
print(df_test_scaled.describe().round(3))
```

```python
df_test_scaled['CountryRegionName'] = df_test['CountryRegionName']
df_test_scaled['CountryRegionName'] = [countries[x] for x in df_test_scaled['CountryRegionName']]
df_test_scaled['Occupation'] = df_test['Occupation']
df_test_scaled['Occupation'] = [occupations[x] for x in df_test_scaled['Occupation']]
df_test_scaled['Education'] = df_test['Education']
df_test_scaled['Education'] = [educations[x] for x in df_test_scaled['Education']]
df_test_scaled['Gender'] = df_test['Gender']
df_test_scaled['Gender'] = [genders[x] for x in df_test_scaled['Gender']]
df_test_scaled['MaritalStatus'] = df_test['MaritalStatus']
df_test_scaled['MaritalStatus'] = [marital_status[x] for x in df_test_scaled['MaritalStatus']]
df_test_scaled['TotalChildren'] = df_test['TotalChildren']
df_test_scaled['NumberCarsOwned'] = df_test['NumberCarsOwned']
df_test_scaled['HomeOwnerFlag'] = df_test['HomeOwnerFlag']
df_test_scaled['NumberChildrenAtHome'] = df_test['NumberChildrenAtHome']
df_test_scaled['AveMonthSpend'] = df_test['AveMonthSpend']
```

```python
test = pd.DataFrame(df_test, columns = ['CustomerID', 'BikeBuyer'])
test['predicted'] = KNN_mod.predict(df_test_scaled)
```

```python
test
```

```python
Correct_predictions = 0
for i in range(test_sample_count):
    if test['BikeBuyer'][i] == test['predicted'][i]:
        Correct_predictions += 1
Accuracy = Correct_predictions/test_sample_count
```
