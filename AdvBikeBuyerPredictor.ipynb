```python
num_cols = ['YearlyIncome', 'AveMonthSpend']
df_scaled = scale(df[num_cols])
df_scaled = pd.DataFrame(df_scaled, columns = num_cols)
print(df_scaled.describe().round(3))
```

```python
countries = {'Australia':0,'United States':1,'Canada':2,'Germany':3,'United Kingdom':4,'France':5}
df_scaled['CountryRegionName'] = df['CountryRegionName']
df_scaled['CountryRegionName'] = [countries[x] for x in df_scaled['CountryRegionName']]
```

```python
occupations = {'Professional':0,'Management':1,'Skilled Manual':2,'Clerical':3,'Manual':4}
df_scaled['Occupation'] = df['Occupation']
df_scaled['Occupation'] = [occupations[x] for x in df_scaled['Occupation']]
```

```python
df_scaled['TotalChildren'] = df['TotalChildren']
df_scaled['NumberCarsOwned'] = df['NumberCarsOwned']
df_scaled['HomeOwnerFlag'] = df['HomeOwnerFlag']
```

```python
df_scaled['BikeBuyer']=df['BikeBuyer']
```

```python
df_scaled
```

```python
df_train = np.asmatrix(df_scaled)
df_train_features = df_train[:,:7]
df_train_labels = np.ravel(df_train[:,7])
```
